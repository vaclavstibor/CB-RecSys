```python
# import settings.config_movie_lens_25m as cfg

dataset_folder = 'ml-25m'
preprocessed_dataset_folder = 'preprocessed_dataset'

display(dataset_folder)
```

```python
import pandas as pd

movies_df = pd.read_csv(dataset_folder+'/movies.csv')

display(movies_df)
```

```python
# Extract the year from the title.
movies_df['year'] = movies_df['title'].str.extract('.*\((\d+)\).*', expand=False).fillna(0).apply(int)

# Seperate the Geners to columns
movies_df = movies_df.join(movies_df.pop('genres').str.get_dummies('|'))

# Clean the table
movies_df = movies_df.drop(['(no genres listed)', 'title'], axis=1)

display(movies_df)
```

```python
tags_df = pd.read_csv(dataset_folder+'/genome-scores.csv')

display(tags_df)
```

```python
# Convert tags to table represents tags vector for each movie

table_tags = tags_df.pivot(index='movieId',
                           columns='tagId',
                           values='relevance')

table_tags = table_tags.reset_index().rename_axis(None, axis=1)

display(table_tags)
```

```python
movies_df = movies_df.merge(table_tags, on='movieId', how='left').fillna(0)

movies_df = movies_df.rename(columns={"movieId": "item"})

#movies_df.set_index("item", inplace=True) 

display(movies_df)
```

```python
movies_df.to_csv(preprocessed_dataset_folder+'/movies.csv', index=False)
movies_df.to_pickle(preprocessed_dataset_folder+'/movies.pkl')
```
